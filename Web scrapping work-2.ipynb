### 14.Python program to scrap 10 images of rayban sunglasses from flipkart

```python
from selenium import webdriver
import os

driver=webdriver.Chrome(executable_path='D:\Softwares\chromedriver.exe')
my_page=driver.get('https://www.flipkart.com/search?q=rayban+glass+for+man&sid=26x&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=rayban+glass+for+man%7CSunglasses&requestId=64c5ec17-81f6-4f25-a55e-1e981c65f164')

def make_directory(dirname):
    current_path=os.getcwd()
    path=os.path.join(current_path, dirname)
    if not os.path.exists(path):
        os.makedirs(path)
        
make_directory("Goggles")
images=driver.find_elements_by_xpath("//img[@class='_3togXc']")

urls=[]
for image in images:
    source=image.get_attribute('src')
    urls.append(source)
    if len(urls)==10:
        break
        
import shutil
import requests

for index, link in enumerate(urls):
    print("Downloading {0} of {1} images".format(index + 1, len(urls)))
    response=requests.get(link)
    with open('Goggles/img_{0}.jpeg'.format(index), "wb") as file:
        file.write(response.content)
```

### 15.program to scrap 20 mouses’ data which includes model name, price from www.amazon.in whose price is less than 500 and make a data frame with 2 columns “model_name”,” price” with the scraped data.

```python
from selenium import webdriver
import os

driver=webdriver.Chrome(executable_path='D:\Softwares\chromedriver.exe')
my_page=driver.get('https://www.amazon.in/s?bbn=1375420031&rh=n%3A976392031%2Cn%3A%21976393031%2Cn%3A1375248031%2Cn%3A1375412031%2Cn%3A1375420031%2Cp_36%3A-50000&qid=1600929128&rnid=1318502031&ref=lp_1375420031_nr_p_36_5')
def make_directory(dirname):
    current_path=os.getcwd()
    path=os.path.join(current_path, dirname)
    if not os.path.exists(path):
        os.makedirs(path)

make_directory("Mouses")
images=driver.find_elements_by_xpath("//img[@class='s-image']")
models=driver.find_elements_by_xpath("//span[@class='a-size-medium a-color-base a-text-normal']")
prices=driver.find_elements_by_xpath("//span[@class='a-price-whole']")

mouse_data={}
mouse_data['image_urls']=[]
mouse_data['models']=[]
mouse_data['prices']=[]

for image in images:
    source=image.get_attribute('src')
    mouse_data['image_urls'].append(source)
    if len(mouse_data['image_urls'])==20:
        break

for model in models:
    mouse_data['models'].append(model.text)
    if len(mouse_data['models'])==20:
        break
        
for price in prices:
    mouse_data['prices'].append(price.text)
    if len(mouse_data['prices'])==20:
        break
        
import shutil
import requests

for index, link in enumerate(mouse_data['image_urls']):
    print("Downloading {0} of {1} images".format(index + 1, len(mouse_data['image_urls'])))
    response=requests.get(link)
    with open('Mouses/img_{0}.jpeg'.format(index), "wb") as file:
        file.write(response.content)
```
